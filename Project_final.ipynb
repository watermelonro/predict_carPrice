## 함수

```python
# 원핫인코딩에서 Brand car 뽑기

def get_Brand_df(data,Brand):
    car = data[data['Brand'] == Brand]
    car.reset_index(drop=True, inplace=True)
    
    return car
```

```python
# 차 가격 예측-LinearRegressor

def get_car_price_lr(data,brand, car, *log):
    X = get_Brand_df(data,brand).drop(['Price', 'Brand'], axis = 1)
    y = get_Brand_df(data,brand)['Price']
    if len(log) > 0:
        y = np.log1p(y)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
    
    lr = LinearRegression()
    lr.fit(X_train, y_train)

    pred_lr = lr.predict(car)
    return pred_lr
```

```python
# 차 가격 예측-RandomForest

def get_car_price_rf(data,brand,car,*log):
    X = get_Brand_df(data,brand).drop(['Price', 'Brand'], axis = 1)
    y = get_Brand_df(data,brand)['Price']
    if len(log) > 0:
        y = np.log1p(y)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)

    rf = RandomForestRegressor()
    rf.fit(X_train, y_train)

    pred_rf = rf.predict(car)
    return pred_rf
```

```python
# 차 가격 예측-LightGBM

def get_car_price_lgbm(data,brand,car,*log):
    X = get_Brand_df(data,brand).drop(['Price', 'Brand'], axis = 1)
    y = get_Brand_df(data,brand)['Price']
    if len(log) > 0:
        y = np.log1p(y)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)

    lgbm = LGBMRegressor()
    lgbm.fit(X_train, y_train)

    pred_lgbm = lgbm.predict(car)
    return pred_lgbm
```

```python
def where_we_should_sell(data,brand):

    X = get_Brand_df(data,brand).drop(['Price', 'Brand'], axis=1)
    y = get_Brand_df(data,brand)['Price']

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=13)

    ols_reg = sm.OLS(y_train, X_train).fit()

    print(f"{brand} should sell your car in {ols_reg.params[7:18].sort_values(ascending=False).index[0].split('_')[1]}")
```

```python
where_we_should_sell(data2,'Volvo')
```

```python
def where_we_should_sell_rf(data,brand):

    X = get_Brand_df(data,brand).drop(['Price', 'Brand'], axis=1)
    y = get_Brand_df(data,brand)['Price']

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=13)

    rf = RandomForestRegressor()
    rf.fit(X_train, y_train)

    value = rf.feature_importances_
    value_s = pd.Series(value, index=X_train.columns)

    print(f"{brand} should sell your car in {value_s[7:18].sort_values(ascending=False).index[0].split('_')[1]}")
```

```python
where_we_should_sell_rf(data2,'Volvo')
```
